```python
# Tema: Gestão de Estoque (E-commerce).

estoque = ["lápis","caneta","caderno"]
estoque.append("borracha")
estoque.pop(0)
print(estoque)

```

```python
estoque_novo = {
    'caneta': {'preço': 5, 'qtd': 10},
    'caderno': { 'preço': 20, 'qtd': 5},
    'borracha': {'preço': 1, 'qtd': 100}
}
print(estoque_novo['caderno'],estoque_novo['borracha'])
```

```python
valor_total_estoque = 0.0

for dados in estoque_novo:

    subtotal = estoque_novo[dados]['preço'] * estoque_novo[dados]['qtd']

    valor_total_estoque += subtotal



print(f"O valor total em estoque é R$ {valor_total_estoque:.2f}")
```
